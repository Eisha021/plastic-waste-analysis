```python
# ============================================================
# Project: Plastic Waste Data Analysis & Visualization
# ============================================================

import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Introduction
# -----------------------------
print("\nPLASTIC WASTE DATA ANALYSIS & VISUALIZATION\n")
print(
    "Plastic pollution is one of the most critical global environmental challenges.\n"
    "Due to widespread plastic consumption and inefficient waste management,\n"
    "large quantities of plastic waste end up in landfills and natural ecosystems.\n"
    "This project analyzes plastic waste generation and recycling rates across\n"
    "selected countries, including Pakistan, to highlight global disparities\n"
    "and emphasize the importance of sustainable waste management solutions.\n"
)

# -----------------------------
# Dataset (Real-world aligned estimates)
# Sources: Our World in Data, World Bank, OECD
# -----------------------------
data = {
    "Country": [
        "United States", "China", "India", "Germany",
        "Brazil", "Japan", "Indonesia", "United Kingdom", "Pakistan"
    ],
    "Plastic_Waste_Mt": [42.0, 59.0, 26.0, 14.5, 11.3, 9.0, 8.2, 7.2, 3.9],
    "Recycled_Mt": [3.3, 18.0, 1.9, 6.8, 1.0, 7.6, 0.7, 3.1, 0.4],
    "Population_M": [331, 1441, 1393, 83, 214, 126, 273, 67, 241]
}

df = pd.DataFrame(data)

# -----------------------------
# Calculations
# -----------------------------
df["Recycling_Rate_%"] = (df["Recycled_Mt"] / df["Plastic_Waste_Mt"]) * 100
df["Mismanaged_Waste_Mt"] = df["Plastic_Waste_Mt"] - df["Recycled_Mt"]
df["Waste_per_Capita_kg"] = (df["Plastic_Waste_Mt"] * 1e9) / (df["Population_M"] * 1e6)

print("DATASET OVERVIEW:\n")
print(df.round(2))

# -----------------------------
# Visualization 1: Total Plastic Waste
# -----------------------------
plt.figure()
plt.bar(df["Country"], df["Plastic_Waste_Mt"])
plt.xticks(rotation=45, ha="right")
plt.title("Annual Plastic Waste Generation by Country")
plt.ylabel("Million Tonnes")
plt.tight_layout()
plt.show()

# -----------------------------
# Visualization 2: Recycled vs Mismanaged Waste
# -----------------------------
plt.figure()
plt.bar(df["Country"], df["Recycled_Mt"], label="Recycled")
plt.bar(
    df["Country"],
    df["Mismanaged_Waste_Mt"],
    bottom=df["Recycled_Mt"],
    label="Mismanaged"
)
plt.xticks(rotation=45, ha="right")
plt.title("Recycled vs Mismanaged Plastic Waste")
plt.ylabel("Million Tonnes")
plt.legend()
plt.tight_layout()
plt.show()

# -----------------------------
# Visualization 3: Recycling Rates
# -----------------------------
plt.figure()
plt.bar(df["Country"], df["Recycling_Rate_%"])
plt.xticks(rotation=45, ha="right")
plt.title("Plastic Recycling Rates by Country (%)")
plt.ylabel("Recycling Rate (%)")
plt.tight_layout()
plt.show()

# -----------------------------
# Key Findings
# -----------------------------
print("\nKEY FINDINGS\n")
print(
    "- High-income countries generate significantly more plastic waste per capita.\n"
    "- Recycling performance varies widely across regions.\n"
    "- Pakistan demonstrates low recycling capacity despite rising plastic use.\n"
    "- A large share of plastic waste remains mismanaged globally.\n"
)

# -----------------------------
# Sustainability Perspective
# -----------------------------
print("SUSTAINABILITY & CIRCULAR ECONOMY CONTEXT\n")
print(
    "The traditional linear economy model of produce-use-dispose has proven\n"
    "ineffective in addressing plastic pollution. A circular economy promotes\n"
    "material reuse, recycling, and waste-to-resource conversion, reducing\n"
    "environmental leakage and conserving resources.\n"
)

# -----------------------------
# Conclusion
# -----------------------------
print("CONCLUSION\n")
print(
    "This analysis highlights critical gaps in global plastic waste management.\n"
    "While developed countries face challenges related to high consumption,\n"
    "developing countries like Pakistan face infrastructure limitations.\n"
    "Addressing plastic pollution requires coordinated efforts focused on\n"
    "sustainable materials and improved recycling systems.\n"
)

# -----------------------------
# Future Work
# -----------------------------
print("FUTURE WORK\n")
print(
    "- Incorporate official datasets directly from global databases\n"
    "- Analyze long-term trends in plastic waste generation\n"
    "- Evaluate the impact of waste management policies\n"
    "- Assess circular economy solutions in developing regions\n"
)

# -----------------------------
# References
# -----------------------------
print("REFERENCES\n")
print(
    "1. Our World in Data. Plastic Pollution.\n"
    "2. World Bank. World Development Indicators.\n"
    "3. OECD. Global Plastics Outlook.\n"
    "4. Geyer et al. (2017). Production, use, and fate of all plastics ever made.\n"
)
```
